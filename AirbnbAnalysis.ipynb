### Null value imputation

```python
#if the data is numerical
#if data is normal => mean imputation
#if data is not normal => median imputation

#if the data is categorical
#mode 
#bfill/ffill
```

```python
# Filling first and last review columns with the most recent date available in the dataset as it is a datetime column
bnb['first_review'].fillna(bnb['first_review'].max(), inplace=True)
bnb['last_review'].fillna(bnb['last_review'].max(), inplace=True)
```

```python
# Filling null values with median imputation for security deposit and cleaning fee column
bnb['security_deposit'].fillna(bnb['security_deposit'].median(), inplace=True)
bnb['cleaning_fee'].fillna(bnb['cleaning_fee'].median(), inplace=True)
```

```python
# Replace empty cells with some values
bnb['name'].replace(to_replace='', value='No Info Available', inplace=True)
bnb['summary'].replace(to_replace='', value='No Summary Available', inplace=True)
bnb['space'].replace(to_replace='', value='No Info Available', inplace=True)
bnb['description'].replace(to_replace='', value='No Description Available', inplace=True)
bnb['neighborhood_overview'].replace(to_replace='', value='No Info Available', inplace=True)
bnb['notes'].replace(to_replace='', value='No Info Available', inplace=True)
bnb['transit'].replace(to_replace='', value='No Info Available', inplace=True)
bnb['access'].replace(to_replace='', value='No Info Available', inplace=True)
bnb['interaction'].replace(to_replace='', value='No Info Available', inplace=True)
bnb['house_rules'].replace(to_replace='', value='No Info Available', inplace=True)
```

```python
def extract_area(address):
    try:
        area = address.get('government_area', 'N/A')
        return area
    except Exception as e:
        return None
```

```python
bnb['area'] = bnb['address'].apply(extract_area)
```

```python
bnb.host[0]
```

```python
def extract_host_name(hostaddress):
    try:
        host_name = hostaddress.get('host_name', 'N/A')
        return host_name
    except Exception as e:
        return None
```

```python
bnb['host_name'] = bnb['host'].apply(extract_host_name)
```

```python
def extract_country(address):
    try:
        country = address.get('country', 'N/A')
        return country
    except Exception as e:
        return None        
```

```python
bnb['country'] = bnb['address'].apply(extract_country)
```

```python
bnb.availability[4]
```

```python
def extract_availability(address):
    try:
        availability = address.get('availability_90', 0)
        return availability
    except Exception as e:
        return None        
```

```python
bnb['availability'] = bnb['availability'].apply(extract_availability)
```

```python
def calculate_average_rating(review_scores):
    keys = [
        'review_scores_accuracy',
        'review_scores_cleanliness',
        'review_scores_checkin',
        'review_scores_communication',
        'review_scores_location',
        'review_scores_value'
    ]
    
    # Collect available scores
    scores = [review_scores[key] for key in keys if key in review_scores]
    
    # Calculate the average if there are available scores
    if scores:
        average_rating = sum(scores) / len(scores)
    else:
        average_rating = 0  
    
    return average_rating
```

```python
bnb['rating'] = bnb['review_scores'].apply(calculate_average_rating)
```

```python
bnb['rating'] = round(bnb['rating'],1)
```

```python
bnb.drop(columns='review_scores', inplace=True)
```

```python
bnb.drop_duplicates()
```

```python
# Identify columns with lists or other unhashable types
for col in bnb.columns:
    if bnb[col].apply(lambda x: isinstance(x, list) or isinstance(x, dict)).any():
        print(f"Column '{col}' contains unhashable types.")
    
```

```python
# Convert lists/dicts to strings to check for duplicates
for col in bnb.columns:
    if bnb[col].apply(lambda x: isinstance(x, list) or isinstance(x, dict)).any():
        bnb[col] = bnb[col].apply(lambda x: str(x) if isinstance(x, (list, dict)) else x)
```

```python
bnb[bnb.duplicated()]
```

```python
# Convert strings back to lists or dicts
columns_to_convert = ['amenities', 'images', 'host', 'address', 'availability', 'reviews']
for col in columns_to_convert:
    if bnb[col].apply(lambda x: isinstance(x, str)).any():
        bnb[col] = bnb[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```
